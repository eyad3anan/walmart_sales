## Which Store has the max number of sales
### Store number 20 has the highest sales

```python
total_sales_for_each_store = data_frame.groupby("Store")["Weekly_Sales"].sum().round().sort_values(ascending = False).reset_index() 
total_sales_for_each_store = pd.DataFrame(total_sales_for_each_store) # Converting into a dataframe
total_sales_for_each_store.head(10) # getting the first 10 rows
```

```python
plt.figure(figsize = (15 , 6))
plt.style.use("ggplot")
plt.bar(total_sales_for_each_store["Store"] , total_sales_for_each_store["Weekly_Sales"])
plt.xlabel("store")
plt.ylabel("Weekly Sales")
plt.xticks(total_sales_for_each_store["Store"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each Month")
plt.show()
```

## Which store has the maxuimum standard deviation or which store it's sales vary alot
### Store number 14 has the highest standard deviation

```python
store_max_std = data_frame.groupby(data_frame["Store"])["Weekly_Sales"].std().sort_values(ascending = False).reset_index()
store_max_std = pd.DataFrame(store_max_std)
store_max_std.head() # getting the first 5 rows
```

```python
plt.figure(figsize = (15 , 6))
plt.style.use("ggplot")
plt.bar(store_max_std["Store"] , store_max_std["Weekly_Sales"])
plt.xlabel("store")
plt.ylabel("Weekly Sales")
plt.xticks(store_max_std["Store"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each Month")
plt.show()
```

```python
super_bowl_dates = ["10-2-2010" , "11-2-2011" , "12-2-2012"]
labour_day_dates = ["7-9-2010" , "9-9-2011" , "10-9-2010"]
thanksgiving_dates = ["23-11-2010" , "25-11-2011" , "26-11-2010"]
christmas_dates = ["28-12-2010" , "30-12-2011" , "31-12-2010"]
super_bowl_dates = pd.to_datetime(super_bowl_dates , format="%d-%m-%Y")
labour_day_dates = pd.to_datetime(labour_day_dates , format="%d-%m-%Y")
thanksgiving_dates = pd.to_datetime(thanksgiving_dates , format="%d-%m-%Y")
christmas_dates = pd.to_datetime(christmas_dates , format="%d-%m-%Y")
```

### mean of sales in each hoilday

```python
super_bowl_sales = data_frame.loc[data_frame["Date"].isin(super_bowl_dates)]["Weekly_Sales"].mean() 
labour_sales = data_frame.loc[data_frame["Date"].isin(labour_day_dates)]["Weekly_Sales"].mean() 
thanksgiving_sales = data_frame.loc[data_frame["Date"].isin(thanksgiving_dates)]["Weekly_Sales"].mean()
christmas_sales = data_frame.loc[data_frame["Date"].isin(christmas_dates)]["Weekly_Sales"].mean()
print(f"mean sales in super bowl holiday is {super_bowl_sales}")
print(f"mean sales in labour bowl holiday is {labour_sales}")
print(f"mean sales in thanksgiving holiday is {thanksgiving_sales}")
print(f"mean sales in christmas holiday is {christmas_sales}")
```

### Provide a monthly and semester view of sales

```python
data_frame["Month"] = data_frame["Date"].dt.month
data_frame["Year"] = data_frame["Date"].dt.year
data_frame.head()
```

```python
monthly_sales = data_frame.groupby(data_frame["Month"])["Weekly_Sales"].sum().round().sort_values(ascending = False).reset_index()
monthly_sales = pd.DataFrame(monthly_sales)
monthly_sales
```

```python
plt.figure(figsize = (12 , 6))
plt.style.use("ggplot")
plt.bar(monthly_sales["Month"] , monthly_sales["Weekly_Sales"])
plt.xlabel("Months")
plt.ylabel("Weekly Sales")
plt.xticks(monthly_sales["Month"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each Month")
plt.show()
```

```python
data_frame["quarter"] = data_frame["Date"].dt.quarter
data_frame['semester'] = data_frame['Year'].astype(str) + ' Q' + data_frame['quarter'].astype(str)
data_frame.head()
```

## getting the sales in each semester

### 2010 Q4 the semester which has the highest sales and 2014 Q4 has the lowest sales

```python
semester_sales = data_frame.groupby(data_frame['semester'])['Weekly_Sales'].sum().round().reset_index()
semester_sales = pd.DataFrame(semester_sales)
semester_sales
```

```python
plt.figure(figsize = (12 , 6))
plt.style.use("ggplot")
plt.bar(semester_sales["semester"] , semester_sales["Weekly_Sales"])
plt.xlabel("semester")
plt.ylabel("Weekly Sales")
plt.xticks(semester_sales["semester"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each semester")
plt.show()
```

```python
all_quarter_sales = data_frame.groupby(data_frame['quarter'])['Weekly_Sales'].sum().sort_values(ascending = False).round().reset_index()
all_quarter_sales = pd.DataFrame(all_quarter_sales)
all_quarter_sales
```

```python
plt.figure(figsize = (12 , 6))
plt.style.use("ggplot")
plt.bar(all_quarter_sales["quarter"] , all_quarter_sales["Weekly_Sales"])
plt.xlabel("quarters")
plt.ylabel("Weekly Sales")
plt.xticks(all_quarter_sales["quarter"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each quarter")
plt.show()
```

## Getting the yearly sales in each year

```python
plt.figure(figsize = (12 , 6))
plt.style.use("ggplot")
plt.bar(yearly_sales["Year"] , yearly_sales["Weekly_Sales"])
plt.xlabel("year")
plt.ylabel("Weekly Sales")
plt.xticks(yearly_sales["Year"])
plt.ticklabel_format(style='plain', axis='y') #
plt.title("Sales Of Each quarter")
plt.show()
```
